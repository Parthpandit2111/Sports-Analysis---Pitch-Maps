```python


# Step 2: Filter data for Shreyas Iyer
iyer_data = df1[df1["Striker"] == "Shreyas Iyer"].copy()

# Step 3: Calculate cumulative runs and strike rate
iyer_data["Cumulative Runs"] = iyer_data["Runs"].cumsum()
iyer_data["Cumulative Balls"] = range(1, len(iyer_data) + 1)
iyer_data["Cumulative Strike Rate"] = (iyer_data["Cumulative Runs"] / iyer_data["Cumulative Balls"]) * 100

# Step 4: Plot Strike Rate Evolution Over Time
plt.figure(figsize=(10, 5))
sns.lineplot(x=iyer_data["Cumulative Balls"], y=iyer_data["Cumulative Strike Rate"], marker="o", color="orange")

# Step 5: Customize Graph
plt.title("Shreyas Iyer - Strike Rate Evolution Over Time", fontsize=14)
plt.xlabel("Balls Faced", fontsize=12)
plt.ylabel("Strike Rate", fontsize=12)
plt.ylim(0, max(iyer_data["Cumulative Strike Rate"]) + 10)  # Adjusting y-axis
plt.grid(True)

# Step 6: Add Text Annotation for Explanation
explanation = "This graph illustrates how Shreyas Iyerâ€™s strike rate changed over his innings. " \
              "Steady rise shows controlled acceleration, while sharp increases indicate aggressive phases. " \
              "Dips represent slow phases or dot balls. The trend highlights his approach to pacing the innings."

plt.figtext(0.5, -0.2, explanation, wrap=True, horizontalalignment='center', fontsize=10)

# Show Graph
plt.show()
```
